```python
# numpy.diag can define either a square 2D array with given values along the
# diagonal or if given a 2D array returns a 1D array that is only the diagonal
# elements. The two array creation functions can be helpful while doing linear
# algebra, as such:

np.diag([1, 2, 3])
```

```python
np.diag([1, 2, 3], 1)
```

```python
a = np.array([[1, 2], [3, 4]])
np.diag(a)
```

```python
# vander(x, n) defines a Vandermonde matrix as a 2D NumPy array. Each column of
# the Vandermonde matrix is a decreasing power of the input 1D array or list or
# tuple, x where the highest polynomial order is n-1. This array creation
# routine is helpful in generating linear least squares models, as such:
np.vander(np.linspace(0, 2, 5), 2)
```

```python
np.vander([1, 2, 3, 4], 2)
```

```python
np.vander((1, 2, 3, 4), 4)
```

# (3) Replicating, joining, or mutating existing arrays

```python
a = np.array([1, 2, 3, 4, 5, 6])
b = a[:2]
b += 1
print('a =', a, '; b =', b)
```

```python
# if we create a two different array then b not collaps with a
a = np.array([1, 2, 3, 4])
b = a[:2].copy()
b += 1
print('a = ', a, 'b = ', b)
```

```python
A = np.ones((2, 2))
B = np.eye(2, 2)
C = np.zeros((2, 2))
D = np.diag((-3, -4))
np.block([[A, B], [C, D]])
```
